## Memoization :
Memoization or memoisation techniques used primarily to speed up computer programs by storing the results of expensive function calls and returning the cached results when the same imputs occur again
used in Dynamic programming

```python
import time # Code without using memoizaton will take more time than using memoization technique

def fib(n) : 
    if n == 1 or n == 0:
        return 1
    else :
        return fib(n - 1) + fib(n - 2)
```

```python
start = time.time()
fib(20)
print(time.time() - start)
```

```python
start = time.time()
fib(40)
print(time.time() - start)
```

```python
# Since time taken when the input was 20 is 0.008 seconds, however, when we increase the input we have seen that time reaches on 95 seconds. 
```

```python
# Using Memoization
def fib2(n, d):
    if n in d:
        return d[n]
    else:
        d[n] = fib2(n - 1, d) + fib2(n - 2, d)
        return d[n]
```

```python
start = time.time()
d = {0:1, 1:1}
fib2(40, d)
print(time.time() - start)
```

```python
# When using memoization technique we have saved pre existing outputs in a dictionary and then we we calculate the output it takes less time than compared to the other code mentioned above
```
